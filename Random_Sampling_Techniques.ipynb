# **RANDOM SAMPLING TECHNIQUES**

Total Population Size = 10,000

Sample Size = 100

Table of Random Numbers : 
> Size = 1000 x 1000 ... Range = 0 to 1,00,000





```python
# Written by Deepthi Tabitha Bennet 

import numpy as np
import pandas as pd
```

# **Simple Random Sampling**

```python
#generates Random Numbers (Population) and stores it in a DataFrame
df_population1 = pd.DataFrame(np.random.randint(0, 100000, size=(1000, 1000)))
print("POPULATION : ")
df_population1
```

```python
print("SAMPLE : ")
a = df_population1
ct = 0

for i in range(0, 100):
    for j in range(0, 100):
        if(a[i][j] < 10000):  # Crietria for selection of the Random Numbers
            if(ct < 100):  # Number of Elements to be chosen = 10
                print(a[i][j], end = " ")
                ct = ct + 1
```

# **Stratified Random Sampling :**

**Splitting the Population of size 10,000 into 5 Categories (internally Homogenous, Externally Heterogenous)**



```python
# Category A - Size 4000
df_population2a = pd.DataFrame(np.random.randint(0, 100000, size=(20, 200)))
print("POPULATION : ")
df_population2a
```

```python
print("SAMPLE : ")
a = df_population2a
ct = 0

for i in range(0, 200):  # no. of columns
    for j in range(0, 20):  # no. of rows
        if(a[i][j] < 10000):  # Crietria for selection of the Random Numbers
            if(ct < 40):  # Number of Elements to be chosen = 10
                print(a[i][j], end = " ")
                ct = ct + 1
```

# **Systematic Random Sampling :**


```python
df_population3 = pd.DataFrame(pd.DataFrame(np.random.randint(0, 100000, size=(1000, 1000))))
print("POPULATION : ")
df_population3
```

```python
k = 10000 / 100  # k = Population Size / Sample Size

print("SAMPLE : ")
a = df_population3
ct = 0

for i in range(0, 1000, int(k)):
    for j in range(0, 1000, int(k)):
        if(ct < 100):  # Number of Elements to be chosen = 100
            print(a[i][j], end = " ")
            ct = ct + 1
```
